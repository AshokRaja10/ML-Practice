```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
input_data = pd.read_csv('Linear_regression/input_data.csv')
y = input_data.Sales
x = input_data.Time
x = x.values.reshape(-1,1)
x_train, x_test, y_train, y_test=train_test_split(x, y,test_size = 0.2, random_state=0)
```

```python
plt.scatter(x_train, y_train, color='red')
plt.title('Time vs Sales')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Sales'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linearRegessionModel = LinearRegression()
linearRegessionModel.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, linearRegessionModel.predict(x_train), color='blue')
plt.title('Time vs Sales Training Data set')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Prediction','Sales'])
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, linearRegessionModel.predict(x_test), color='blue')
plt.title('Time vs Sales Test Data set')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Prediction','Sales'])
plt.show()
```

```python
print( " x1 = "+ str(linearRegessionModel.coef_))
```

```python
x_new = pd.Series([45,67,24,78])
x_new = x_new.values.reshape(-1,1)
linearRegessionModel.predict(x_new)
```

```python
print("Training accuracy: " + str(r2_score(y_train, linearRegessionModel.predict(x_train))))
print("Testing  accuracy: " + str(r2_score(y_test , linearRegessionModel.predict(x_test ))))
```
