```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import operator
%matplotlib inline
```

```python
x = 2 - 3 * np.random.normal(0, 0.7, 60)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-5, 5, 60)
x = x.reshape(-1, 1) 

plt.scatter(x, y, color='magenta')
plt.title('Time vs Sales Training Traning Data set')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Sales'])
plt.show()
```

```python
linearRegessionModel = LinearRegression()
linearRegessionModel.fit(x, y)
```

```python
plt.scatter(x, y, color='magenta')
plt.plot(x, linearRegessionModel.predict(x), color='blue')
plt.title('Time vs Sales Training Data set')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Prediction','Sales'])
plt.show()
```

```python
mean_squared_error(y,linearRegessionModel.predict(x))
```

```python
r2_score(y,linearRegessionModel.predict(x))
```

```python
polynomial_features = PolynomialFeatures(degree=5)
x_poly = polynomial_features.fit_transform(x)
```

```python
polynomialRegessionModel = LinearRegression()
polynomialRegessionModel.fit(x_poly, y)
```

```python
plt.scatter(x, y, color='magenta')

sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,polynomialRegessionModel.predict(x_poly)), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='blue')

plt.title('Time vs Sales Training Data set')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend(['Prediction','Sales'])
plt.show()
```

```python
mean_squared_error(y, polynomialRegessionModel.predict(x_poly))
```

```python
r2_score(y, polynomialRegessionModel.predict(x_poly))
```
